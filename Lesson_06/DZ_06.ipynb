Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
#перечень всех документов
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def request_document_number():
    num_doc_request = input('Введите номер документа: ')
    return num_doc_request

def owner_by_number(num_doc_owner): # поиcк владельца документа по его номеру
    for element in documents:
        if num_doc_owner == element['number']:
            return element['name']

def print_p():    # вывод информации по кнопке 'p'
    document_number_p = request_document_number()
    if owner_by_number(document_number_p) == None:
        print('Документ не найден в базе')
    else:
        print('Владелец документа: ', owner_by_number(document_number_p))
        
def find_shelf(num_doc_find): # поиск номера полки по номеру документа
    for key, value in directories.items():
        if num_doc_find in value:
            return key

def print_s():    # вывод информации по кнопке 's'
    document_number_s = request_document_number()
    if find_shelf(document_number_s) == None:
        print('Документ не найден в базе')
    else:
        print('Документ хранится на полке: ', find_shelf(document_number_s))
        
def info_all_docs(): #вывод информации по кнопке 'l'
    for element in documents:
        print('№: ', element['number'], ', тип: ', element['type'], end='')
        print(', владелец: ', element['name'], ', полка хранения: ', find_shelf(element['number']))

def list_shelf(): # вывод списка полок
    counter = 1
    for row in directories:
        if counter !=len(directories):
            print(row, end=', ')
        else:
            print(row, end='.')
        counter +=1
    print()

def request_shelf_number():
    num_shelf_request = input('Введите номер полки: ')
    return num_shelf_request
        
def add_shelf(): #Пользователь по команде "ads" может добавить новую полку
    num_shelf_add = request_shelf_number()
    if num_shelf_add in directories:
        print('Такая полка уже существует. Текущий перечень полок: ', end='')
        list_shelf()
    else:
        directories.setdefault(num_shelf_add,[])
        print('Полка добавлена. Текущий перечень полок: ', end='')
        list_shelf()

def del_shelf(): #Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
    num_shelf_del = request_shelf_number()
    if num_shelf_del not in directories:
        print('Такой полки не существует. Текущий перечень полок: ', end='')
        list_shelf()
    elif len(directories.get(num_shelf_del)) != 0:
        print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ', end='')
        list_shelf()
    elif len(directories.get(num_shelf_del)) == 0:
        del(directories[num_shelf_del])
        print('Полка удалена. Текущий перечень полок: ', end='')
        list_shelf()

while True:
    command = input('Введите команду ') 
    if command == 'p': # Пользователь по команде "p" может узнать владельца документа по его номеру
        print_p()
    elif command == 's': #Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
        print_s()
    elif command == 'l': # Пользователь по команде "l" может увидеть полную информацию по всем документам
        info_all_docs()
    elif command == 'as': # Пользователь по команде "as" может добавить новую полку¶
        add_shelf()
    elif command == 'ds': # Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
        del_shelf()
    elif command == 'q':
        break
````

№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

````python
#перечень всех документов
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def del_document_shelf(document_number_del_sh): # удаление номера документа c полки, полку находит сам
    directories[find_shelf(document_number_del_sh)].remove(document_number_del_sh)
    
def del_document_documents(document_number_del_doc): # удаление документа из документов по номеру документа
    index = 0
    for el in documents:
        if document_number_del_doc in el.values():
            del(documents[index])
        index += 1
    
def add_document_shelf(shelf_number_add, document_number_add): # добавление номера документа в указанную полку
    directories[shelf_number_add].append(document_number_add)

def add_document_documents(document_type_add_f, num_doc_add_f, document_owner_add_f):
    documents.append(dict(zip(['type','number','name'],[document_type_add_f,num_doc_add_f,document_owner_add_f])))
    
def request_document_number(): # запрос и получение номер документа
    num_doc_request = input('Введите номер документа: ')
    return num_doc_request

def request_shelf_number():    # запрос и получение номера полки
    num_shelf_request = input('Введите номер полки: ')
    return num_shelf_request

def request_document_type(): # запрос и получение типа документа
    document_type = input('Введите тип документа: ')
    return document_type

def request_document_owner(): # запрос и получение владельца документа
    document_owner = input('Введите владельца документа: ')  
    return document_owner

def request_storage_shelf(): # запрос и получение полка для хранения
    storage_shelf = input('Введите полку для хранения: ')    
    return storage_shelf

def owner_by_number(num_doc_owner): # поиcк владельца документа по его номеру
    for element in documents:
        if num_doc_owner == element['number']:
            return element['name']

def find_shelf(num_doc_find): # поиск номера полки по номеру документа
    for key, value in directories.items():
        if num_doc_find in value:
            return key

def print_p():    # вывод информации по кнопке 'p'
    document_number_p = request_document_number() # запрос и получение номер документа
    if owner_by_number(document_number_p) == None: # вызов функции поиcк владельца документа по его номеру и условие
        print('Документ не найден в базе')
    else:
        print('Владелец документа: ', owner_by_number(document_number_p)) # вызов функции поиcк владельца документа по его номеру и условие
        
def print_s():    # вывод информации по кнопке 's'
    document_number_s = request_document_number() # запрос и получение номер документа
    if find_shelf(document_number_s) == None: # поиск номера полки по номеру документа и условие
        print('Документ не найден в базе')
    else:
        print('Документ хранится на полке: ', find_shelf(document_number_s)) # поиск номера полки по номеру документа
        
def info_all_docs(): #вывод информации по кнопке 'l'
    print('Текущий список документов:')
    for element in documents:
        print('№: ', element['number'], ', тип: ', element['type'], end='')
        print(', владелец: ', element['name'], ', полка хранения: ', find_shelf(element['number']))

def list_shelf(): # вывод списка полок
    counter = 1
    print('Текущий список полок:')
    for row in directories:
        if counter !=len(directories):
            print(row, end=', ')
        else:
            print(row, end='.')
        counter +=1
    print()
    
def list_doc_shelf():
    for key, value in directories.items():
        print('Номер полки - ',key, '. Документы:', value)

def add_shelf(): #Пользователь по команде "as" может добавить новую полку
    num_shelf_add = request_shelf_number() # вызов функции запрос и получение номера полки
    if num_shelf_add in directories:
        print('Такая полка уже существует. ', end='')
        list_shelf() # вызов функции вывод списка полок
    else:
        directories.setdefault(num_shelf_add,[])
        print('Полка добавлена. ', end='')
        list_shelf() # вызов функции вывод списка полок

def del_shelf(): #Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
    num_shelf_del = request_shelf_number() # вызов функции запрос и получение номера полки
    if num_shelf_del not in directories:
        print('Такой полки не существует. ', end='')
        list_shelf() # вызов функции вывод списка полок
    elif len(directories.get(num_shelf_del)) != 0:
        print('На полке есть документы, удалите их перед удалением полки. ', end='')
        list_shelf() # вызов функции вывод списка полок
    else:
        del(directories[num_shelf_del]) # вызов функции удалении полки
        print('Полка удалена. ', end='')
        list_shelf() # вызов функции вывод списка полок

def move_document(): # Пользователь по команде "m" может переместить документ с полки на полку
    document_number_move = request_document_number() # запрос и получение номер документа
    shelf_number_move = request_shelf_number()    # запрос и получение номера полки
    if shelf_number_move not in directories: 
        print('Такой полки не существует.', end='')
        list_shelf()
    elif find_shelf(document_number_move) == None:
        print('Документ не найден в базе.')
        info_all_docs()
    else:
        del_document_shelf(document_number_move)
        add_document_shelf(shelf_number_move, document_number_move)
        print('Документ перемещен.')
        info_all_docs()
        list_doc_shelf()
            
def del_document():# Пользователь по команде “d” может удалить документ из данных
    document_number_del = request_document_number() # запрос и получение номер документа
    if find_shelf(document_number_del) == None:
        print('Документ не найден в базе.')
        info_all_docs()
    else:
        del_document_documents(document_number_del)
        del_document_shelf(document_number_del)
        print('Документ удален.')
        info_all_docs()
        list_doc_shelf()

def add_document(): # Пользователь по команде “ad” может добавить новый документ в данные
    document_type_add = request_document_type() # запрос и получение типа документа
    num_doc_add = request_document_number() # запрос и получение номер документа
    document_owner_add = request_document_owner() # запрос и получение владельца документа
    storage_shelf_add = request_storage_shelf() # запрос и получение полка для хранения
    if storage_shelf_add not in directories: 
        print('Такой полки не существует. Добавьте полку командой as.')
        info_all_docs()
    else:
        add_document_documents(document_type_add, num_doc_add, document_owner_add)
        add_document_shelf(storage_shelf_add, num_doc_add) # добавление номера документа в указанную полку
        print('Документ добавлен. ', end='')
        info_all_docs()
        list_doc_shelf()

while True:
    command = input('Введите команду ') 
    if   command == 'p':  # Пользователь по команде "p" может узнать владельца документа по его номеру
        print_p()
    elif command == 's':  # Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
        print_s()
    elif command == 'l':  # Пользователь по команде "l" может увидеть полную информацию по всем документам
        info_all_docs()
        list_shelf()
        list_doc_shelf()
    elif command == 'as': # Пользователь по команде "as" может добавить новую полку¶
        add_shelf()
    elif command == 'ds': # Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
        del_shelf()
    elif command == 'q':  # выход из программы
        break
    elif command == 'ad': # Пользователь по команде "ad" может добавить новый документ в данные
        add_document()
    elif command == 'd':  # Пользователь по команде "d" может удалить документ из данных
        del_document()
    elif command == 'm':  # Пользователь по команде "m" может переместить документ с полки на полку
        move_document()
````

##### Как правильно задавать вопросы аспирантам, преподавателям и коллегам
Прежде чем задать вопрос, попробуйте найти ответ в интернете. Навык самостоятельного поиска информации — один из важнейших. Каждый практикующий специалист любого уровня делает это ежедневно.

Сформулируйте вопрос по алгоритму:  
1) Что я делаю?  
2) Какого результата я ожидаю?  
3) Как фактический результат отличается от ожидаемого?  
4) Что я уже попробовал сделать, чтобы исправить проблему?  

По возможности прикрепите к вопросу скриншоты либо ссылки на код. Не выкладывайте все решение, оставляйте только проблемный и воспроизводимый участок кода.
